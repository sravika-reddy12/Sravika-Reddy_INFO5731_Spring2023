Question 2 (20 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# You code here (Please add comments in the code):
from six.moves import urllib
import pandas as pd    # importing pandas

t1 = [] #url github
url = "https://raw.githubusercontent.com/sravika-reddy12/Sravika-Reddy_INFO5731_Spring2023/main/sra.txt"
tf = urllib.request.urlopen(url)  
#Use this function to decode the url and retrieve its info. It has a predetermined purpose.#
for i in tf: # for loop
#The data in the url is decoded and put into standard format using this statement.
    dec_data = i.decode("utf-8").replace('\r\n', '')    
    if dec_data: # if loop
        t1.append(dec_data)       
#adding a list to the decoded data
df = pd.DataFrame (t1, columns = ['Text'])   
#Pandas is used to transform a list to a data frame.
print(df)
#printing statement
```

```python
#To find the number of special characters present in the list

def spchar(x):
    red = 0
    for i in range(len(x)): #for loop
        if not((x[i].isalpha()) and (x[i].isdigit())):   # if statement
            red = red + 1
    return red  #return
# df
df['Num of spec char'] = df['Text'].apply(lambda y: spchar(y))
df
```

```python
#To find Number of Sentences in a data frame
#df
df['Num of sentences'] = df['Text'].apply(lambda y: len(str(y).split(".")))
df
```

```python
#To find Number of words in a data frame
#df
df['Num of Words'] = df['Text'].apply(lambda y: len(str(y).split(" ")))
df
```

```python
#To find Number of characters in data frame
#df
df['Num of Char'] = df['Text'].str.len()
df
```

```python
#To find the number of numerics present in the list

df['Num of num'] = df['Text'].apply(lambda x: len([x for x in x.split() if x.isdigit()])) 
#here we are considering it to be a seperate numeric if it is seperated with a space using the split function
df
```

```python
#To find number of uppercase words in a list

df['Num of upper case words'] = df['Text'].apply(lambda x: len([x for x in x.split() if x.isupper()])) 
#using split function we are finding the words with completely upper case
df
```

```python
df.to_csv('data.csv',index=True)
```

Question 3 (10 points): Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)." Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
from sklearn.feature_selection import chi2
from sklearn.feature_extraction.text import CountVectorizer

# Converting the text data into a matrix 
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Text'])

# Calculate the Chi-Square statistic 
#p-values for each feature
chi_scores, p_values = chi2(X, df['Text'])

#  feature names from the CountVectorizer object
feature_names = vectorizer.get_feature_names_out()

# Combine the feature names with their corresponding scores and sort in descending order
scores = list(zip(feature_names, chi_scores))
scores.sort(key=lambda x: x[1], reverse=True)

# Printing the features ranked by importance
print("Rank\tFeature\t\t\t\tChi-Square Score")
for i, score in enumerate(scores):
    print(f"{i+1}\t{score[0]:<20}\t{score[1]:.2f}")
```

```python
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
# Load the pre-trained BERT model
model = SentenceTransformer('bert-base-nli-mean-tokens')
text_embeddings = model.encode(df['Text'])
query = "text data"
# Encode the query using the BERT model
query_embedding = model.encode(query)
# Calculating the cosine similarity between the query and each text in the data
similarity_scores = cosine_similarity(query_embedding.reshape(1, -1), text_embeddings)
# Rank the similarity scores in descending order
sorted_indices = np.argsort(similarity_scores, axis=1)[0][::-1]
# Print the text data sorted by similarity
sorted_text_data = df.iloc[sorted_indices]['Text']
#printing statement
print(sorted_text_data)
```

```python
import torch
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import urllib.request

# Load pre-trained BERT model
model = SentenceTransformer('bert-base-nli-mean-tokens')

# Load text data
#url github
url = "https://raw.githubusercontent.com/sravika-reddy12/Sravika-Reddy_INFO5731_Spring2023/main/sra.txt"
tf = urllib.request.urlopen(url)

t1 = []
for i in tf: #for loop
    dec_data = i.decode("utf-8").replace('\r\n', '')
    if dec_data: #if statement
        t1.append(dec_data)

# Convert text data to embeddings
embeddings = model.encode(t1)
#embeddings
# Define a query and convert it to an embedding
query = "text similarity"
query_embedding = model.encode([query])

# Calculate cosine similarity between query and text data
similarity_scores = cosine_similarity(query_embedding, embeddings)

# Rank the similarity scores in descending order
indices = similarity_scores.argsort()[0][::-1]

# Print the text data ranked by similarity scores
for index in indices:
    print("Text {}: {:.4f}".format(index+1, similarity_scores[0][index]))
```
